## Функция перевода + и - в 1 и 0
#### в некоторых предметах кроме + и - встречаются цифры (они останутся без изменений)

```python
def plusToNum(s):
    st=''
    for i in s:
        if i == '+':
            st+='1'
        elif i == '-': 
            st +='0'
        else:
            st += i
    arr = [int(i) for i in st]
    return arr
```
