# Компьютерная модель движения тела на языке Python

### На основе формальной модели, описывающей движение тела, брошенного под углом к горизонту, создадим компьютерную модель с использованием системы программирования Jupyter notebook

## Проект "Бросание мячика в стенку" на языке Python

```python
import math
```

```python
V0 = 18 # начальная скорость
A = 35  # угол бросания мяча
S = 25  # расстояние до стены
H = 1   # высота стены
G = 9.81
```

```python
# Попадание в стену
L = S * math.tan(A*math.pi / 180) - G * S**2 / (2 * (V0 * math.cos(A * math.pi/180)))
```

```python
print(L)
```

```python
if L < 0:
    print('Недолет')
elif L > H:
    print('Перелет')
else:
    print('Попадание')
```

```python
step = [i/10 for i in range(0, 28, 2)]
print(step)
#V0 * cos(radian(A)) * step $B$1*SIN(RADIANS($B$2))*A5-4,9*A5*A5
x = [V0*math.cos(math.radians(A))*i for i in step]
y = [V0*math.sin(math.radians(A))*i - 4.9*i*i for i in step]
print(x)
print(y)
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
plt.plot(x,y)
plt.show()
```
