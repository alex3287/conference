# Библиотека Pandas
### Рссмотрим объекты Series и DataFrame

* Импортируем библиотеку Pandas и библиотку random
* Создаем два списка заполненные случайными числами

```python
import pandas as pd
from random import randint as rnd

A1 = [rnd(-10, 10) for i in range(10)]
B1 = [rnd(-10, 10)/10 for i in range(10,20)]
```

### Создаем и выводим на экран объект DataFrame

```python
test = pd.DataFrame({'First': A, 'Second':B})
test
```

## Проект "Бросание мячика" на языке Python

### Объявим необходимые переменные
**Необходимые формулы**
* =A1/(V0\*COS(RADIANS(A1)))
* V0\*SIN(RADIANS(A1))\*t1[i]-4,9\*t1[i]\*t1[i]

```python
import math
import pandas as pd

V0 = 18 # (м/с) начальная скорость
A1 = 20  # (град) угол бросания мяча 
A2 = 40  # (град) угол бросания мяча 
A3 = 60  # (град) угол бросания мяча 

x = pd.Series([i for i in range(0,36,2)])

t1 = pd.Series([round(i/(V0*math.cos(math.radians(A1))),1) for i in x], index=x)
y1 = pd.Series([round((V0*math.sin(math.radians(A1))*i-4.9*i*i) ,1) for i in t1], index=x)
t2 = pd.Series([round(i/(V0*math.cos(math.radians(A2))),1) for i in x], index=x)
y2 = pd.Series([round((V0*math.sin(math.radians(A2))*i-4.9*i*i) ,1) for i in t2], index=x)
t3 = pd.Series([round(i/(V0*math.cos(math.radians(A3))),1) for i in x], index=x)
y3 = pd.Series([round((V0*math.sin(math.radians(A3))*i-4.9*i*i) ,1) for i in t3], index=x)
data = pd.DataFrame({'t1':t1,'y1':y1,'t2':t2,'y2':y2,'t3':t3,'y3':y3})
data
```

### Выборка данных

```python
data2 = data[['y1', 'y2', 'y3']]
data2
```
